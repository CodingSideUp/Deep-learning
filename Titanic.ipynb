```python
columns_to_drop = ["Cabin", "Ticket", "PassengerId", "Name"]
df = df.drop(columns=columns_to_drop)
df.head()
```

```python
import numpy as np
age_mean = df.Age.mean()
replacements = {"Sex": {"male": 1, "female":0}, "Age": {np.NaN: age_mean}}
df.replace(replacements, inplace=True)
df.head()
```

```python
print("Length before dropping:", len(df))
df.dropna(inplace=True)
df.reset_index()
print("Length after dropping:", len(df))
```

```python
df = pd.get_dummies(df)
df.head()
```

```python

from sklearn.model_selection import train_test_split

df_train, df_valid = train_test_split(df, test_size=0.1)

X_train = df_train.drop(columns=["Survived"])
y_train = df_train["Survived"]

X_valid = df_valid.drop(columns=["Survived"])
y_valid = df_valid["Survived"]

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_valid shape:", X_valid.shape)
print("y_valid shape:", y_valid.shape)
```

```python
import numpy as np

class Perceptron:
   def step_func(self, z):
      """Activation function: Step function"""
      return np.where(z >= 0, 1, 0)

   def fit(self, X, y, rate=0.001, epoch=500):
      """Fit training data
      X : Training vectors, X.shape : [#samples, #features]
      y : Target values, y.shape : [#samples]
      """
      n_samples, n_features = X.shape

      # Initialize weights and bias
      self.weight = np.zeros(n_features)
      self.bias = 0

      for e in range(epoch):
         n_miss = 0  # Number of misclassifications

         # Random shuffle the data
         perm = np.random.permutation(n_samples)

         for xi, target in zip(X[perm], y[perm]):
            # Calculate y_hat (prediction using step function)
            y_hat = self.step_func(np.dot(xi, self.weight) + self.bias)

            # Calculate delta W if target != y_hat (misclassification)
            if target != y_hat:
               delta_w = rate * (target - y_hat)  # Change in weights

               # Update weight and bias if misclassification occurs
               self.weight += delta_w * xi
               self.bias += delta_w

               # Increase misclassifications number if applied
               n_miss += 1

         # Print the number of misclassifications every 50 epochs
         if e % 50 == 0:
            print(f"Epoch {e} - Wrong prediction: {n_miss}")

      print("Finished training. Lost:", n_miss)

   def predict(self, X):
      """Return class label after unit step"""
      return self.step_func(np.dot(X, self.weight) + self.bias)

```

```python
# Ensure numeric types
X_train = X_train.astype(np.float64)
y_train = y_train.astype(np.int64)

p = Perceptron()
p.fit(X_train.to_numpy(), y_train.to_numpy(), 0.001, 500)
```
